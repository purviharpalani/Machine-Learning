```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("UCI_Credit_Card.csv")
df
```

```python
for i in range(25):
    print("The number of null values in column ",i,"are:")
    nullcheck=df.iloc[:,i].isnull().sum()
    print(nullcheck)
```

```python
Y=df['default.payment.next.month']
Y
```

```python
X=df.drop(['ID','default.payment.next.month'],axis='columns')
X
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
X.iloc[:,12:24]=scaler.fit_transform(X.iloc[:,12:24])
X
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,train_size=0.8,random_state=7)
```

```python
X_train
```

```python
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
lr.fit(X_train,Y_train)
Y_predicted=lr.predict(X_test)
print(Y_predicted[0:5])
print(np.array(Y_test[0:5]))
print("The accuracy:")
print(lr.score(X_test,Y_test))
```

```python
from sklearn import metrics
matrix=metrics.confusion_matrix(Y_predicted,Y_test)
sns.heatmap(matrix,annot=True,fmt='0.1f')
plt.xlabel("Predicted values")
plt.ylabel("Test values")
plt.title("Confusion Matrix for Logistic Regression")
plt.show()
```

```python
from sklearn.naive_bayes import GaussianNB
nb=GaussianNB()
nb.fit(X_train,Y_train)
Y_predicted=nb.predict(X_test)
print("The first five predicted values are:")
print(Y_predicted[0:5])
print("The first five test values are:")
print(np.array(Y_test[0:5]))
print("The accuracy of NaiveBayesClassifier is:")
print(nb.score(X_test,Y_test))
```

```python
from sklearn import metrics
matrix=metrics.confusion_matrix(Y_predicted,Y_test)
sns.heatmap(matrix,annot=True,cmap="Blues",fmt='0.1f')
plt.xlabel("Predicted values")
plt.ylabel("Test values")
plt.title("Confusion Matrix for Logistic Regression")
plt.show()
```

```python
from sklearn.model_selection import cross_val_score
lr.fit(X,Y)
cross_val_score(lr,X,Y,cv=10)
np.average(cross_val_score(lr,X,Y,cv=10))
```

```python
lr.fit(X,Y)
cross_val_score(nb,X,Y,cv=10)
np.average(cross_val_score(nb,X,Y,cv=10))
```
